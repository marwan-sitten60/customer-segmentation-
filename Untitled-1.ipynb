## Create Model 

### Way 1

```python
SD=StandardScaler()
x_data=SD.fit_transform(df)
```

```python
model=kmeans.KMeans(n_clusters=9,init='k-means++',max_iter=300,n_init=10,random_state=0) # Create a KMeans model with 5 clusters
```

```python
model.fit(df) # Fit the model to the standardized data
```

```python
model.inertia_ # Get the inertia of the model, which is the sum of squared distances to the nearest cluster center
```

```python
inertias = []
for i in range(1, 11):
    model = kmeans.KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    model.fit(df)
    inertias.append(model.inertia_)
plt.plot(range(1, 11), inertias, 'ro-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

#### Observations:
- As the number of clusters increases from 1 to 10, the inertia decreases sharply at first.
- The point where the decrease in inertia starts to level off (the "elbow") occurs at **k = 4**.
- This indicates that 3 clusters may be the optimal choice for the dataset, as adding more clusters beyond this point yields diminishing returns in reducing inertia.


```python
model=kmeans.KMeans(n_clusters=4,init='k-means++',max_iter=300,n_init=10,random_state=0)
model.fit(df) # Fit the model to the standardized data
Pre=model.predict(df) # Predict the cluster labels for the data
```

```python
df['k_Meam']=Pre
```
